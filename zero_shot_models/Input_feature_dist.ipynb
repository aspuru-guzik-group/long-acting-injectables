```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

# Histogram distributions of features and target

```python
datafile = "Dataset_3.0_15_feat.xlsx"

df = pd.read_excel(datafile)

X = df.drop(['Experimental_index','DP_Group','Release'],axis='columns')
Y = df['Release']
G = df['DP_Group']

#check with and without standard scalar
stdScale = StandardScaler().fit(X)
X_scale=stdScale.transform(X)
```

```python
# original dataframe
X
```

### Reconstrcut datafrme for X_scaled

```python
names = X.columns.to_list() # get column names
```

```python
# convert np.array to dataframe
df_2 = pd.DataFrame(data=X_scale[1:,1:],    # values
             index=X_scale[1:,0],    # 1st column as index
             columns=X_scale[0,1:])  # 1st row as the column names

df_2
```

```python
# rest index
df_2 = df_2.reset_index(level=0)
df_2
```

```python
# rest columns to first row
df_3 = pd.DataFrame(np.vstack([df_2.columns, df_2]))
df_3
```

```python
# add column names
df_3.columns = names
df_3
```

```python
# impute 0,0
df_3.loc[0][0] = 0.192489
df_3
```

```python
# libraries & dataset
import seaborn as sns
import matplotlib.pyplot as plt
# set a grey background (use sns.set_theme() if seaborn version 0.11.0 or above) 
sns.set(style="darkgrid")


sns.histplot(data=df, x="Polymer_MW", kde=True,  bins=20)
plt.show()
```

### Figures to print

```python
for name in names:
    
    # Histogram of all the drug polymer groups in the dataset
    f, fig = plt.subplots(figsize=(12, 6))

    sns.set(style="white")

    fig = sns.histplot(data=df, x=name, kde=True, color="red", bins=10)

    fig.lines[0].set_color('black')

    # Title
    fig.axes.set_title("Distribution of "+str(name)+" in the Dataset", fontsize=14, color="black", weight="bold")

    # Title - x-axis/y-axis
    fig.set_xlabel(name, fontsize=12, color='black', weight='bold')
    fig.set_ylabel("Count", fontsize=12, color='black', weight='bold')

    # x-axis and y-axis tick color
    fig.tick_params(colors='black', which='both')  # 'both' refers to minor and major axes

    # add tick marks on x-axis or y-axis
    fig.tick_params(bottom=False, left=True)
    
    plt.tight_layout()
    
    filename = name.replace("/"," ") # remove "/" from string name

    plt.savefig('Figures/Histograms/Feature_'+str(filename), dpi=600, format = 'png', transparent=True, bbox_inches='tight')

    #plt.show()
```
